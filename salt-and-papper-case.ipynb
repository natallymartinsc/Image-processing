```python
#imports
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
#Loading the image
image = cv2.imread('Fig0510.tif') # Load image
```

This image it's an example of salt-and-papper noise because as we can see, the pixels are disrupted bu extreme values - white and black, causing a lot of dots on the image.

I applied 4 differents filters to see which one it's the best for the salt-and-papper case. The filters are:


1.   Average filter
2.   Gaussian filter
3.   Median filter

After the results, i'll discuss the results according the filter

```python
#Applying some filters to see the best result
#we are using a 5x5 kernel
average_filtered = cv2.blur(image, (5, 5))
gaussian_filtered = cv2.GaussianBlur(image, (5, 5), 0)
median_filtered = cv2.medianBlur(image, 5)
```

```python
#Here i'm displaying the original salt-and-papper image
plt.title('Original image')
plt.imshow(image, cmap='gray')
plt.axis('off')
```

```python
#here i'm plotting the filtered images
plt.figure(figsize=(20, 4))

plt.subplot(1, 4, 1)
plt.title('Original image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title('Average filter')
plt.imshow(average_filtered, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title('Gaussian filter')
plt.imshow(gaussian_filtered, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title('Median filter')
plt.imshow(median_filtered, cmap='gray')
plt.axis('off')


plt.tight_layout()
plt.show()
```

#Results:

The Median filter it's effective for this type of noise - salt-and-papper.
According to Tan and Jiang, 2019, the median filter it's very effective to remove impulse noise like the salt and pepper noise because the median filter has the principle to replace the gray level of each pixel by the median of the gray levels in a neighborhood of the pixels.



*   Average filter: this filter simply smooths our image but it's still possible to see the dots;
*   Gaussian filter: we applied the gaussian blur filter that smooths our image too, but it's still possible to see the dots;
*   Median filter: the filter that we applied that had the best result;
