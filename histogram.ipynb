```python
#imports
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

#Low contrast case

```python
#Reading the image and converting to the gray scale
im1 = cv2.imread('Fig0316.tif')
img = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
```

```python
#histogram equalization
histoEqualization = cv2.equalizeHist(img)
```

#What is histogram equalization?
It's a technique which the goal is to redistribute the pixel intensity values, in other words, improve the contrast of the image

```python
plt.figure(figsize=(12, 6)) #to display side by side

# plotting the histogram of our original image
plt.subplot(1, 2, 1)
plt.title("Original histogram")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.hist(img.ravel(), 256, [0, 256])
plt.xlim([0, 256])

#plotting the histogram of the equalized image
plt.subplot(1, 2, 2)
plt.title("Equalized histogram")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.hist(histoEqualization.ravel(), 256, [0, 256])
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

#Discuss about the original histogram

As we can see on the original histogram, there are a lot of pixels that have high intensity values close to 255, causing a low contrast image

```python
plt.figure(figsize=(12, 6)) #to display side by side

# plotting the histogram of our original image
plt.subplot(1, 2, 1)
plt.title("Original histogram")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.hist(img.ravel(), 256, [0, 256])
plt.xlim([0, 256])
```

#Results:

As we can see, after the histogram equalization the contrast it was improved. The right image it has better contrast compared to our original image on the left.


*   A highly concentrated pixel region has a problem for correction



```python
#displaying the images side by side
figure, axarr = plt.subplots(1,2, figsize=(20,20))
axarr[0].imshow(im1, cmap='gray')
axarr[1].imshow(histoEqualization, cmap='gray')
```

#High Contrast Case

```python
#Reading the image and converting to the gray scale
im2 = cv2.imread('Fig0316HighContrast.tif')
image = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
```

```python
histoEqualization2 = cv2.equalizeHist(image)
```

#Discuss

As we can see on the original histogram, there are a lot of pixels close to 0, causing a high contrast image

```python
plt.figure(figsize=(12, 6)) #to display side by side

# plotting the histogram of our original image
plt.subplot(1, 2, 1)
plt.title("Original histogram")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.hist(image.ravel(), 256, [0, 256])
plt.xlim([0, 256])

#plotting the histogram of the equalized image
plt.subplot(1, 2, 2)
plt.title("Equalized histogram")
plt.xlabel("Intensity")
plt.ylabel("Number of pixels")
plt.hist(histoEqualization2.ravel(), 256, [0, 256])
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
#displaying the images side by side
figure, axarr = plt.subplots(1,2, figsize=(20,20))
axarr[0].imshow(im2, cmap='gray')
axarr[1].imshow(histoEqualization2, cmap='gray')
```
